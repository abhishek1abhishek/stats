```python
df2
```

```python
df2['name'].isnull().value_counts()
```

```python
df2['age'].isnull().value_counts()
```

```python
df2['marks'].isnull().value_counts()
```

```python
df2.dropna()
```

```python
df2.dropna(axis='index',how='any')      #any means it will delete any row/index which contains NONE value
```

```python
df2.dropna(axis='index',how='all')         #all means it will delete any row/index which contains NONE value in every respective columns
```

```python
df2.dropna(axis='columns',how='any')      #here it will drop column
```

```python
df2.dropna(axis='index',how='any',subset=['name'])
```
