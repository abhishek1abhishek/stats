```python
import seaborn as sns
s3=sns.load_dataset('mpg')
```

```python
s3
```

```python
sns.lineplot(x='horsepower',y='weight',data=s3,hue='origin',markers=['o','^','>'],palette='ocean',dashes=False)
```

```python
sns.displot(s3['horsepower'],bins=[0,50,100,150,300],kde=True,color='k',rug=True)
#kernel density estimator
```

```python
#barplot
sns.barplot(x=s3.horsepower,y=s3.origin)
```

```python
markers = {'USA': '*', 'Europe': 'o', 'Japan': '^'}
sizes = (10, 100, 90)
sns.scatterplot(x='horsepower',y='weight',data=s3,hue='origin'
                ,style='origin',sizes=sizes,legend=True)
```

```python
marker={'usa':'*','japan':'o','europe':'>'}
sns.scatterplot(x='horsepower',y='weight',data=s3,
                hue='origin',style='origin',sizes=['10','10','10000'],palette='magma',
                alpha=.8,markers=marker)
```

```python
x = 'horsepower'
y = 'weight'
hue = 'origin'
sizes = [10, 100, 90]  # List specifying the size for each category
markers = {'USA': '*', 'Europe': 'o', 'Japan': '^'}  # Dictionary specifying marker style for each category

# Create the scatter plot with custom sizes and markers
sns.scatterplot(x=x, y=y, data=s3, hue=hue, style=hue, s=100)  # s=100 sets the default size for all points

# Customize marker style for each category
for category, marker in markers.items():
    plt.scatter([], [], marker=marker, label=category, color='gray', s=100)
```
