```python
#this is something different

import numpy as np
import matplotlib.pyplot as plt

# Example data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create histogram
counts, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black')

# Calculate bin width
bin_width = bins[1] - bins[0]

# Convert counts to PDF by normalizing
pdf = counts / (np.sum(counts) * bin_width)

# Plot PDF
plt.plot(bins[:-1], pdf, color='red', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Probability Density Function (PDF)')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate example age data
ages = np.random.randint(18, 80, size=1000)  # Generating 1000 random ages between 18 and 80

# Create a histogram to visualize the age distribution
counts, bins, _ = plt.hist(ages, bins=30, density=True, alpha=0.5, color='blue', edgecolor='black')

# Calculate the PDF
bin_width = bins[1] - bins[0]  # Calculate the width of each bin
pdf = counts / (np.sum(counts) * bin_width)

# Plot the PDF
plt.plot(bins[:-1], pdf, color='red', linewidth=2)
plt.title('Probability Density Function (PDF) of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```
